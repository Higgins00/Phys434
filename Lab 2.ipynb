```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
from scipy import special
```

$\text{I chose to use  X =0.6 and Y = 2}$

```python
mu1d = 0.6
k1d = np.arange(0,51)

ppmf1d = stats.poisson.pmf(k1d,mu1d)
```

```python
fig, ax = plt.subplots(1,5,figsize=(20,10))
ppmf = ppmf1d

ax[0].bar(k1d,ppmf1d,width=1)
ax[0].set_ylabel("Probability")
ax[0].set_xlabel("Total Background Counts")
ax[0].set_xlim(-.5,20)

for j in np.arange(1,5):
    ppmf = np.convolve(ppmf,ppmf1d,mode ='full')
    ax[j].bar(np.arange(0,len(ppmf)),ppmf,width=1,label = "{0:d} Days".format(j+1))
    ax[j].set_xlim(-.5,20)
    ax[j].set_xlabel("Total Background Counts")
    ax[j].legend()
```

We can see the mean of the distrubition gets larger, but the distribution is becoming more symmetric about the mean.

### B).

```python
ppmf = ppmf1d

for j in np.arange(1,5):
    ppmf = np.convolve(ppmf1d,ppmf,'full')

fig,ax = plt.subplots(1,2,figsize=(12,5))

ax[0].bar(np.arange(0,len(ppmf)),ppmf,width =1)
ax[0].set_xlim(-.5,100)
ax[0].set_xlabel('Total Background Counts')
ax[0].set_ylabel('Probability');

ax[1].bar(np.arange(0,len(ppmf)),ppmf,width =1)
#ax[1].set_xlim(-.5,50)
ax[1].set_xlabel('Total Background Counts')
ax[1].set_ylabel('Log Scale Probability')
ax[1].set_yscale('log');
```

The graph of the log plot of the probability looks like that of a poisson distribution. The fact that it remained a Poisson distribution after just a few convolutions makes sense conceptually because any poisson distributed random variable that is  added to another poisson distributed variable should maintain their quality of being poisson except for when a large number of them are summed together because then the central limit theorem states that the pmf will approach a Guassian. This can be explained mathematically as $\Large{\sum pois(\lambda_1) +pois(\lambda_2) = \sum pois(\lambda_1 + \lambda_2)}$

$$ = \sum \left( \frac{e^{-\lambda_1}\lambda_1^x}{x!} + \frac{e^{-\lambda_2}\lambda_2^y}{y!} \right)$$

$$= e^{-(\lambda_1 + \lambda_2)}\sum \left( \frac{\lambda_1^x}{x!} + \frac{\lambda_2^y}{y!} \right)$$

$$ \text{in our case x = y = k} $$

$$\text{New Poisson} =  \frac{e^{-(\lambda_1 + \lambda_2)}(\lambda_1 + \lambda_2)^k}{k!}$$

This new added Poison has the same function as the original Poisson $\frac{e^{-\lambda}\lambda^k}{k!}$ except the means of the two poisson's are now summed together

```python
fig, ax = plt.subplots(11,2,figsize=(10,25))
ppmf = ppmf1d

for i in np.arange(0,10):
    ppmf = np.convolve(ppmf,ppmf1d,mode ='full')
    ax[i][0].bar(np.arange(0,len(ppmf))/(i+2),ppmf,width=1,label = "{0:d} Days Average".format(i+2))
    ax[i][0].set_xlim(-.5,10)
    ax[i][0].set_xlabel("Average Background Counts Per Day")
    ax[i][0].legend()
        
    ax[i][1].bar(np.arange(0,len(ppmf))/(i+2),ppmf,width=1,label = "{0:d} Days Average".format(i+1))
    ax[i][1].set_yscale('log')
    
ppmf = ppmf1d
for j in np.arange(0,60):
    ppmf = np.convolve(ppmf1d,ppmf,'full')

ax[10][0].bar(np.arange(0,len(ppmf))/(61),ppmf,width =1, label = "61 Days Average")
ax[10][0].set_xlim(-.5,10)
ax[10][0].set_xlabel('Average Background Counts')
ax[10][0].set_ylabel('Probability');
ax[10][0].legend()

ax[10][1].bar(np.arange(0,len(ppmf))/j+2,ppmf,width =1)
ax[10][1].set_xlabel('Total Background Counts')
ax[10][1].set_ylabel('Log Scale Probability')
ax[10][1].set_yscale('log');
```

We can see that as we average over more and more days the distribution has a smaller and smaller spread of counts(visible in the Log plots). This makes sense from part B since we can see that as the sum of the means gets larger the larger average counts will become less and less probable. This makes sense with the Central Limit Theorem because as the sum gets infintely large the distribution will approach a normal distribution with a center at k = X = 0.6 in my case.(We can see this in the last graph with the upside down parabola starting to take shape)

### D).

I choose to use N = 61 for the number of days. 

This makes it so Y * N = 122 gamma rays over the 61 days.

I will calculate the pmf for 61 days by convolving with itself 60 times. Then I will sum the probabilities of all the values in the array $\geq$ 122 counts. This will give me the probability of the 122 gamma rays occuring from just the background. I will then find the sigma value these by looking at a guassian distribution and integrating from the right side until I get the probability and this will allow me to get my sigma value.

```python
#ppmf found in previous cell for 61 summed days and still initialized
probability = ppmf[122:len(ppmf):1].sum()
sigma= stats.norm.isf(probability)
print("There is a {0:.2f} sigma probability that my background would have had a count of 122 gamma rays over 61 days.".format(sigma))
```

## Problem 4:

### A).

```python
# 5 sigma probability
sigma5prob = stats.norm.sf(5)

sigmav1 = stats.norm.isf(sigma5prob,scale=2)

print('In order to get a 5 Sigma probability of detection in Version 1 we would need to detect a brightness of {0:.1f}'.format(sigmav1))
```

### B).

```python
trialsfactor = 10000
sigmav2 = stats.norm.isf(sigma5prob/trialsfactor,scale=2)

print('In order to get a 5 Sigma probability of detection in Version 2 we would need to detect a brightness of {0:.1f}'.format(sigmav2))
```

### C).

The signal needs to be 31% larger or 1.31 times the value of the Version 1 5sigma value in order to be signifigant to the probability of 5sigma. For Version 1 we get that the brightness needs to be at least 10 to be considered significant, but for Version 2 we need it to be 13.1 in order to be considered significant. This can be explained by our probability plots since once we get to large values of brightness the probability drops of very steeply (that -x^2 plot in the log view). This means that our trials number is low enough that it doesn't have a large effect on the far end of the distribution and doesn't end up contributing much to changing the probability of having a signal-free event at larger brightness numbers.

### D).

```python
ordersofmag = np.arange(0,11)
sigmasatdifferenttrials = stats.norm.isf(sigma5prob/(10**ordersofmag),scale=2)

plt.bar(ordersofmag,sigmasatdifferenttrials)
plt.ylabel(r'Brightness needed for a 5$\sigma$ detection')
plt.xlabel(r'Trials Factor Number $10^x$')
plt.title('Looking at Different trials factors from 1,10,100,1000,10000,...');
```

Above I plotted the Brightness needed for a $5\sigma$ probability of detection as a function of the trials factor. The x-axis shows what power of 10 was used as the trial factor, ie. 0 corresponds to 1, 1 to 10, and so on. We can see that at $10^{10}$ trials the brightness is only about 1.2 times the value at 10k trials. This shows that even at large trial numbers our brightness sensitivity threshold for $5/sigma$ only goes up by about a factor of 2.
